```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score, mean_squared_error

# load the cleaned data from phase 1
df = pd.read_csv("cleaned_employee_dataset.csv")

# separating features (x) and target (y)

x = df.drop('Salary_INR', axis=1)
y = df['Salary_INR']

# 80-20 split for train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# finding the top 8 features using forward selection
print("Picking the top 8 features...")
lin_model = LinearRegression()
sfs = SequentialFeatureSelector(lin_model, n_features_to_select=8, direction='forward')

x_train_sel = sfs.fit_transform(x_train, y_train)
x_test_sel = sfs.transform(x_test)


# Model Training & Testing

print("\n--- Models Output ---")

# 1. Linear Regression
lr = LinearRegression()
lr.fit(x_train_sel, y_train)
lr_preds = lr.predict(x_test_sel)

lr_r2 = r2_score(y_test, lr_preds)
lr_rmse = np.sqrt(mean_squared_error(y_test, lr_preds))
print(f"Linear Reg -> R2: {lr_r2:.4f} | RMSE: ₹ {lr_rmse:.2f}")

# 2. Ridge Regression
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(x_train_sel, y_train)
ridge_preds = ridge_model.predict(x_test_sel)

ridge_r2 = r2_score(y_test, ridge_preds)
ridge_rmse = np.sqrt(mean_squared_error(y_test, ridge_preds))
print(f"Ridge Reg -> R2: {ridge_r2:.4f} | RMSE: ₹ {ridge_rmse:.2f}")

# 3. Lasso Regression
lasso_model = Lasso(alpha=0.5)
lasso_model.fit(x_train_sel, y_train)
lasso_preds = lasso_model.predict(x_test_sel)

lasso_r2 = r2_score(y_test, lasso_preds)
lasso_rmse = np.sqrt(mean_squared_error(y_test, lasso_preds))
print(f"Lasso Reg -> R2: {lasso_r2:.4f} | RMSE: ₹ {lasso_rmse:.2f}")
```
