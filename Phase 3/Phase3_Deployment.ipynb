```python
import pandas as pd
import pickle
import os

# 1. Load the previously trained final model and saved features
if not os.path.exists('final_salary_model.pkl'):
    print("Error: 'final_salary_model.pkl' not found.")
    print("Please run 'Phase3_Final_Training.ipynb' first to generate the model file.")
else:
    with open('final_salary_model.pkl', 'rb') as f:
        saved_data = pickle.load(f)
            
    final_model = saved_data['model']
    expected_features = saved_data['feature_names']
    
    print("Model loaded successfully.\n")

    print("-" * 50)
    print("EMPLOYEE SALARY PREDICTOR - DEPLOYMENT PHASE")
    print("-" * 50)
    print("Please enter the following details:\n")

    # 2. Ask the user for input dynamically
    user_input = {}

    for feature in expected_features:
        val = input(f"Enter value for '{feature}': ").strip()
        user_input[feature] = [float(val)]

    # 3. Convert the user's answers into a DataFrame table
    input_df = pd.DataFrame(user_input)

    # 4. Display the entered data
    print("\n" + "-" * 50)
    print("DATA ENTERED:")
    print("-" * 50)
    for feature, val in user_input.items():
        print(f" - {feature}: {val[0]}")
    print("-" * 50)

    # 5. Make the final prediction
    predicted_salary = final_model.predict(input_df)[0]

    print(f"\nPREDICTED SALARY: ₹ {predicted_salary:,.2f}")
    print("-" * 50)
```
